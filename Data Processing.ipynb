## Feature Scaling | Normalization

[Feature Scaling](https://en.wikipedia.org/wiki/Feature_scaling#:~:text=Feature%20scaling%20is%20a%20method,during%20the%20data%20preprocessing%20step.) is a method used to normalize the range of independent variables or features of data. In data processing, it is also known as data normalization and is generally performed during the data preprocessing step.

Since the range of values of raw data varies widely, in some machine learning algorithms, objective functions will not work properly without normalization. For example, many classifiers calculate the distance between two points by the Euclidean distance. If one of the features has a broad range of values, the distance will be governed by this particular feature. Therefore, the range of all features should be normalized so that each feature contributes approximately proportionately to the final distance.

Another reason why feature scaling is applied is that gradient descent converges much faster with feature scaling than without it.

It's also important to apply feature scaling if regularization is used as part of the loss function (so that coefficients are penalized appropriately).

### Standard Normalization
The standard normalization scales values to between 0 and 1. You can think of this as a mapping function.
$$
\begin{gather}
j = \text{The feature or column in the dataset} \\
max = \text{Max value of the x dataset} \\ \\
x_j = \frac{x_j}{x_{max}}
\end{gather}
$$

```python
from matplotlib import pyplot as plt
import numpy as np
from utilities.utils import normalize_std
%matplotlib inline

x = [15, 9.8, 45, 90, 105, 274.6, 500, 600, 800]
y = np.zeros((9, 1))
fig, axs = plt.subplots(1,2, figsize=(10,1))
fig.suptitle('Standard Normalization', y=1.2)

x_normalized = normalize_std(x)

axs[0].plot(x, y)
axs[0].set_xlabel("Raw Data")
axs[0].set_yticks([])
axs[1].plot(x_normalized, y)
axs[1].set_xlabel("Normalized Data (Standard)")
axs[1].set_yticks([])
```

### Mean Normalization
The mean normalization scales values to between -1 and 1, centered around 0,0. Again, think of this as a mapping function but can allow for negative values.

$$
\begin{gather}
j = \text{The feature or column in the dataset} \\
\mu = \text{Mean of the entire dataset} \\
min = \text{Min value of the x dataset} \\
max = \text{Max value of the x dataset} \\ \\
x_j = \frac{x_j - \mu_j}{x_{max} - x_{min}}
\end{gather}
$$

```python
from matplotlib import pyplot as plt
import numpy as np
from utilities.utils import normalize_mean
%matplotlib inline

x = [15, 9.8, 45, 90, 105, 274.6, 500, 600, 800]
y = np.zeros((9, 1))
fig, axs = plt.subplots(1,2, figsize=(10,1))
fig.suptitle('Mean Normalization', y=1.2)

x_normalized = normalize_mean(x)

axs[0].plot(x, y)
axs[0].set_xlabel("Raw Data")
axs[0].set_yticks([])
axs[1].plot(x_normalized, y)
axs[1].set_xlabel("Normalized Data (Mean)")
axs[1].set_yticks([])
```

### Z-Score Normalization
The z-score normalization will have a mean of zero and a standard deviation of 1, giving a normal distribution.

$$
\begin{gather}
i = \text{A single example from the dataset} \\
j = \text{The feature or column in the dataset} \\
\mu = \text{Mean of the entire dataset} \\
\sigma = \text{The standard deviation} \\ \\
x^{(i)}_j = \frac{x^{(i)}_j - \mu_j}{\sigma_j}
\end{gather}
$$

```python
from matplotlib import pyplot as plt
import numpy as np
from utilities.utils import normalize_zscore
%matplotlib inline

x = [15, 9.8, 45, 90, 105, 274.6, 500, 600, 800]
y = np.zeros((9, 1))
fig, axs = plt.subplots(1,2, figsize=(10,1))
fig.suptitle('Z-score Normalization', y=1.2)
    
x_normalized = normalize_zscore(x)

axs[0].plot(x, y)
axs[0].set_xlabel("Raw Data")
axs[0].set_yticks([])
axs[1].plot(x_normalized, y)
axs[1].set_xlabel("Normalized Data (Z-Score)")
axs[1].set_yticks([])
```
