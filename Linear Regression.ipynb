# Linear Regression - Supervised Learning

Linear regression analysis is used to predict the value of a variable based on the value of another variable. The value you want to predict is called the "dependent variable", while the variable you are using to predict the other variable is called the "independent variable".

The technique can be described by plotting data on a graph and fitting a "best fit line" against that data. This "line" will represent the prediction space you will use when determining the predicted value given the input features.

The simplest function one can use to describe this is:

$$
f_w,_b(\vec{x}) = \vec{w}\bullet\vec{x} + b
$$
Further described as:
$$
f_\vec{w},_b(\vec{x}) = w_0x_0 + w_1x_1 + w_nx_n + b
$$

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

# Initialize figure
fig, ax = plt.subplots(figsize = (5, 5))
ax.set(xlim=(0, 10), ylim=(0, 1000))

# Training Data
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [15, 9.8, 45, 90, 105, 274.6, 500, 600, 800]

# Plot the training data
ax.scatter(x, y, c="black")

# Plot the regression line - Linear
def plotLine(x, y, degree, color, label):
    coefficient = np.polyfit(x, y, deg=degree)
    f_wb = np.poly1d(coefficient)
    xx = np.linspace(0,10,1000)
    yy = f_wb(xx)
    ax.plot(xx, yy, c=color, label=label)

plotLine(x, y, degree=1, color="blue", label="Linear")
plotLine(x, y, degree=2, color="teal", label="2-dimension polynomial")
plotLine(x, y, degree=3, color="orange", label="3-dimension polynomial")
plotLine(x, y, degree=4, color="red", label="4-dimension polynomial")

plt.legend()
```

### Additional Polynomial Functions

Obviously a straight line may not be the best fit for a given dataset. One can choose multiple dimensions of polynomial functions in order to best fit the data. For example, you can use [Cubic Functions](https://en.wikipedia.org/wiki/Cubic_function), [Quadradic Functions](https://en.wikipedia.org/wiki/Quadratic_function), [Squareroot Functions](https://en.wikipedia.org/wiki/Square_root), and more.

**Linear Function**
$$
f_\vec{w},_b(\vec{x}) = w_0x_0 + w_1x_1 + w_nx_n + b
$$

**Quadradic Function**
$$
f_\vec{w},_b(\vec{x}) = w_1x + w_2x^2 + b
$$

**Cubic Function**
$$
f_\vec{w},_b(\vec{x}) = w_1x + w_2x^2 + w_3x^3 + b
$$

**Squareroot Function**
$$
f_\vec{w},_b(\vec{x}) = w_1x + w_2\sqrt{x}
$$
