# Linear Regression - Supervised Learning

Linear regression analysis is used to predict the value of a variable based on the value of another variable. The value you want to predict is called the "dependent variable", while the variable you are using to predict the other variable is called the "independent variable".

The technique can be described by plotting data on a graph and fitting a "best fit line" against that data. This "line" will represent the prediction space you will use when determining the predicted value given the input features.

The simplest function one can use to describe this is:

$$
f_w,_b(\vec{x}) = \vec{w}\bullet\vec{x} + b
$$
Further described as:
$$
f_\vec{w},_b(\vec{x}) = w_0x_0 + w_1x_1 + w_nx_n + b
$$

### Additional Polynomial Functions

Obviously a straight line may not be the best fit for a given dataset. One can choose multiple dimensions of polynomial functions in order to best fit the data. For example, you can use [Cubic Functions](https://en.wikipedia.org/wiki/Cubic_function), [Quadradic Functions](https://en.wikipedia.org/wiki/Quadratic_function), [Squareroot Functions](https://en.wikipedia.org/wiki/Square_root), and more.

**Linear Function**
$$
f_\vec{w},_b(\vec{x}) = w_0x_0 + w_1x_1 + w_nx_n + b
$$

**Quadradic Function**
$$
f_\vec{w},_b(\vec{x}) = w_1x + w_2x^2 + b
$$

**Cubic Function**
$$
f_\vec{w},_b(\vec{x}) = w_1x + w_2x^2 + w_3x^3 + b
$$

**Squareroot Function**
$$
f_\vec{w},_b(\vec{x}) = w_1x + w_2\sqrt{x}
$$

## Underfitting & Overfitting

When fitting the best fit line onto your data, there's a possibility of "underfitting" or "overfitting" your regression line to your data. 

**"Underfitting"** implies that the trend line doesn't represent the best fit line to your data. When you go to make a prediction, an underfit model will make inaccurate predictions. For example, when looking at the following scatter plot, you will see that the data follows a near exponential curve. When we fit a linear regression line the error rates for the data on the ends of the data ranges are much higher than if we had fit a curve instead.

**Overfitting** implies that the trend line fits too closely to the training data. While this may seem like a good thing, the problem is that the model doesn't generalize when making predictions on new data. 

The goal with these lines is to find the balance between accuracy and being able to generalize well. 

**Note:** See the squared cost error value for each dimension of the polynomial function. In this case, the cost is the lowest in the Cubic permutation.

```python
from matplotlib import pyplot as plt
import numpy as np
from utilities.utils import plotAx
%matplotlib inline

fig, axs = plt.subplots(2,2, figsize=(6,6))

# Training Data
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [15, 9.8, 45, 90, 105, 274.6, 500, 600, 800]

# Plot figures
plotAx(axs[0,0], data=(x,y), title='Linear (1d)', dimension=1)
plotAx(axs[0,1], data=(x,y), title='Binomial (2d)', dimension=2)
plotAx(axs[1,0], data=(x,y), title='Cubic (3d)', dimension=3)
plotAx(axs[1,1], data=(x,y), title='Quadradic (4d)', dimension=4)

fig.suptitle('Cost VS Polynomial Dimensions', fontsize=16)
plt.tight_layout()
```
